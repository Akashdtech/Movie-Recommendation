```python
from zipfile import ZipFile
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Dataset - https://www.kaggle.com/datasets/gargmanas/movierecommenderdataset?select=movies.csv

with ZipFile('/content/Movie Recommendation.zip', 'r') as zip:
  zip.extractall()
  print('Done')
```

```python
df = pd.read_csv('/content/movies.csv')
df.head(10)
```

```python
display(df.columns)
display(df.shape)
df.isnull().sum()
```

```python
df['combined_features'] = df['genres'] + ' ' + df['title']
df.head(10)
```

```python
df.shape
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['combined_features'])
```

```python
cos_similarity = cosine_similarity(tfidf_matrix)
```

```python
def get_movie_recommendations(movie_title, cos_similarity=cos_similarity):
  movie_index = df[df['title'] == movie_title].index[0]
  similarity_scores = list(enumerate(cos_similarity[movie_index]))
  sorted_similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
  top_movies = sorted_similarity_scores[1:11]
  recommended_movies = [df['title'].iloc[movie[0]] for movie in top_movies]
  return recommended_movies
```

```python
movie = 'Hitman (2007)'
recommendations = get_movie_recommendations(movie)
print(recommendations)
```
